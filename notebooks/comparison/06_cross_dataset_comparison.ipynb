```python
# ── Table 1: Classification datasets — F1 and AUC ───────────────
classification_df = df_all[df_all["Task"] == "Classification"].copy()

table_clf = classification_df.pivot_table(
    index=["Dataset", "Model", "Type"],
    values=["CV Macro F1", "CV F1 std", "AUC"],
    aggfunc="first"
).round(3)

print("=" * 70)
print("TABLE 1: Classification Model Comparison (CV Macro F1 and AUC)")
print("=" * 70)
print(table_clf.to_string())
```

```python
# ── Table 2: Regression datasets — R² and RMSE ──────────────────
regression_df = df_all[df_all["Task"] == "Regression"].copy()

table_reg = regression_df.pivot_table(
    index=["Dataset", "Model", "Type"],
    values=["CV R²", "CV RMSE"],
    aggfunc="first"
).round(3)

print("=" * 70)
print("TABLE 2: Regression Model Comparison (CV R² and CV RMSE)")
print("=" * 70)
print(table_reg.to_string())
```

```python
# ── Figure 1: Best model per dataset per type ────────────────────
# Shows the best-performing model of each type across all datasets

best_models = df_all.copy()

# For classification use CV Macro F1, for regression use CV R²
best_models["primary_metric"] = best_models.apply(
    lambda r: r["CV Macro F1"] if r["Task"] == "Classification"
              else r["CV R²"],
    axis=1
)

best_per_type = (
    best_models.groupby(["Dataset", "Type"])["primary_metric"]
    .max()
    .reset_index()
)

# Pivot for grouped bar chart
pivot = best_per_type.pivot(
    index="Dataset", columns="Type", values="primary_metric"
)

# Reorder datasets by level
dataset_order = [
    "MH Tech",
    "Student Depression",
    "Canadian Survey",
    "WHR Gallup"
]
pivot = pivot.reindex(dataset_order)

fig, ax = plt.subplots(figsize=(12, 6))

x = np.arange(len(pivot.index))
width = 0.22
colors = {"Baseline": "#4C72B0", "Hybrid": "#DD8452", "BN": "#55A868"}

for i, model_type in enumerate(["Baseline", "Hybrid", "BN"]):
    if model_type in pivot.columns:
        bars = ax.bar(
            x + i * width,
            pivot[model_type],
            width,
            label=model_type,
            color=colors[model_type],
            alpha=0.85
        )
        for bar in bars:
            height = bar.get_height()
            if not np.isnan(height):
                ax.text(
                    bar.get_x() + bar.get_width() / 2.,
                    height + 0.005,
                    f"{height:.3f}",
                    ha="center", va="bottom", fontsize=8
                )

ax.set_xticks(x + width)
ax.set_xticklabels(pivot.index, fontsize=10)
ax.set_ylabel("Best CV Score\n(Macro F1 for classification, R² for regression)",
              fontsize=10)
ax.set_title(
    "Best Model Performance by Type and Dataset\n"
    "(Classification: Macro F1 | Regression: R²)",
    fontsize=12
)
ax.legend(title="Model Type", fontsize=9)
ax.set_ylim(0, 1.05)
ax.axhline(y=0, color="black", linewidth=0.5)

# Add level annotations
for i, dataset in enumerate(dataset_order):
    level = "Micro" if dataset != "WHR Gallup" else "Macro"
    ax.text(
        i + width, -0.07,
        f"({level})",
        ha="center", fontsize=8, color="gray"
    )

plt.tight_layout()
plt.savefig("../../outputs/shap/comparison_best_by_type.png",
            dpi=150, bbox_inches="tight")
plt.show()
print("Saved: comparison_best_by_type.png")
```

```python
# ── Figure 2: Hybrid vs Baseline improvement across datasets ─────

baseline_best = (
    best_models[best_models["Type"] == "Baseline"]
    .groupby("Dataset")["primary_metric"].max()
    .reset_index()
    .rename(columns={"primary_metric": "baseline_best"})
)

hybrid_best = (
    best_models[best_models["Type"] == "Hybrid"]
    .groupby("Dataset")["primary_metric"].max()
    .reset_index()
    .rename(columns={"primary_metric": "hybrid_best"})
)

improvement = baseline_best.merge(hybrid_best, on="Dataset")
improvement["delta"] = improvement["hybrid_best"] - improvement["baseline_best"]
improvement = improvement.set_index("Dataset").reindex(dataset_order)

fig, ax = plt.subplots(figsize=(9, 5))

colors_delta = [
    "#55A868" if d > 0 else "#C44E52"
    for d in improvement["delta"]
]

bars = ax.barh(
    improvement.index,
    improvement["delta"],
    color=colors_delta,
    alpha=0.85
)

for bar, val in zip(bars, improvement["delta"]):
    ax.text(
        val + (0.001 if val >= 0 else -0.001),
        bar.get_y() + bar.get_height() / 2,
        f"{val:+.3f}",
        ha="left" if val >= 0 else "right",
        va="center",
        fontsize=9
    )

ax.axvline(x=0, color="black", linewidth=1)
ax.set_xlabel(
    "Hybrid improvement over best baseline\n"
    "(Macro F1 for classification, R² for regression)",
    fontsize=10
)
ax.set_title(
    "Hybrid Model Gain over Best Baseline\nby Dataset",
    fontsize=12
)
ax.invert_yaxis()
plt.tight_layout()
plt.savefig("../../outputs/shap/comparison_hybrid_gain.png",
            dpi=150, bbox_inches="tight")
plt.show()
print("Saved: comparison_hybrid_gain.png")
```

```python
# ── Figure 3: BN vs best ML model ───────────────────────────────

ml_best = (
    best_models[best_models["Type"].isin(["Baseline", "Hybrid"])]
    .groupby("Dataset")["primary_metric"].max()
    .reset_index()
    .rename(columns={"primary_metric": "ml_best"})
)

bn_perf = (
    best_models[best_models["Type"] == "BN"]
    .groupby("Dataset")["primary_metric"].max()
    .reset_index()
    .rename(columns={"primary_metric": "bn_score"})
)

bn_comparison = ml_best.merge(bn_perf, on="Dataset")
bn_comparison = bn_comparison.set_index("Dataset").reindex(dataset_order)

x = np.arange(len(bn_comparison.index))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 5))

ax.bar(x - width/2, bn_comparison["ml_best"],
       width, label="Best ML", color="#4C72B0", alpha=0.85)
ax.bar(x + width/2, bn_comparison["bn_score"],
       width, label="Bayesian Network", color="#55A868", alpha=0.85)

for i, (ml, bn) in enumerate(zip(
    bn_comparison["ml_best"], bn_comparison["bn_score"]
)):
    ax.text(i - width/2, ml + 0.005, f"{ml:.3f}",
            ha="center", va="bottom", fontsize=8)
    ax.text(i + width/2, bn + 0.005, f"{bn:.3f}",
            ha="center", va="bottom", fontsize=8)

ax.set_xticks(x)
ax.set_xticklabels(bn_comparison.index, fontsize=10)
ax.set_ylabel(
    "CV Score\n(Macro F1 for classification, R² for regression)",
    fontsize=10
)
ax.set_title(
    "Best ML Model vs Bayesian Network\nby Dataset",
    fontsize=12
)
ax.legend(fontsize=9)
ax.set_ylim(0, 1.05)

# Add note on BN evaluation asymmetry
ax.text(
    0.99, 0.02,
    "Note: BN scores evaluated on full dataset\n"
    "(not CV) — not directly comparable to ML CV scores",
    transform=ax.transAxes,
    ha="right", va="bottom",
    fontsize=7, color="gray",
    style="italic"
)

plt.tight_layout()
plt.savefig("../../outputs/shap/comparison_bn_vs_ml.png",
            dpi=150, bbox_inches="tight")
plt.show()
print("Saved: comparison_bn_vs_ml.png")
```

```python
# ## Cross-Dataset Model Comparison
#
# ### Overview
# Four datasets were evaluated across three model families — baseline
# ML, hybrid ML, and Bayesian Networks — using consistent evaluation
# protocols. Classification tasks report CV Macro F1 and AUC;
# regression tasks report CV R² and CV RMSE. BN scores are evaluated
# on the full observed dataset and are not directly comparable to ML
# CV scores due to evaluation asymmetry.
#
# ### RQ1: Performance at Individual vs Societal Level
# Performance varies substantially across datasets and reflects both
# the modeling level and the quality of the predictive signal:
#
# - MH Tech (Micro, Classification): best ML = 0.488 Macro F1.
#   The hardest dataset in the study — a three-class outcome with
#   moderate class imbalance and noisy workplace survey features.
# - Student Depression (Micro, Classification): best ML = 0.841
#   Macro F1. The strongest individual-level result, driven by
#   a concentrated signal in academic pressure and financial stress.
# - Canadian Survey (Micro, Regression): best ML = 0.423 R².
#   Moderate predictive power for a continuous wellbeing outcome
#   with heterogeneous health and socioeconomic predictors.
# - WHR Gallup (Macro, Regression): best ML = 0.850 R².
#   The strongest result overall, reflecting clean empirical
#   structure in societal-level indicators.
#
# The macro-level dataset (WHR Gallup) achieves approximately twice
# the R² of the comparable micro-level regression dataset (Canadian
# Survey, R² = 0.423). This supports the thesis argument that
# societal-level wellbeing indicators have more consistent and
# predictable relationships with aggregate outcomes than individual-
# level psychological predictors do with individual outcomes.
# However, this comparison should be interpreted carefully —
# the datasets differ not only in level but also in construct
# (happiness score vs life satisfaction) and predictor quality.
#
# ### RQ1: Hybrid vs Baseline Models
# The hybrid gain plot reveals a consistent pattern across all four
# datasets — hybrid models never substantially outperform the best
# baseline:
# - MH Tech: +0.018 Macro F1 (largest gain, but still modest)
# - Canadian Survey: +0.001 R² (negligible)
# - Student Depression: -0.001 Macro F1 (marginal loss)
# - WHR Gallup: -0.010 R² (slight loss)
#
# This finding is robust and consistent: for wellbeing prediction
# tasks, feature selection and ensemble stacking add implementation
# complexity without meaningful predictive gain. The signal
# structure in these datasets is stable enough that well-configured
# baseline models capture it fully. This is consistent with the
# findings of Breiman (2001) on the diminishing returns of model
# complexity when the data generating process is relatively stable.
#
# ### RQ2: BN vs ML Performance
# The BN vs ML comparison shows a dataset-dependent pattern:
# - Student Depression: BN = 0.826 vs ML = 0.841 — closest gap
#   in the study. The BN performs nearly as well as ML despite
#   using a generative rather than discriminative approach,
#   suggesting the causal structure is clean and well-specified.
# - MH Tech: BN = 0.471 vs ML = 0.488 — small gap, comparable
#   performance. The BN captures the core signal despite the
#   noisier dataset.
# - WHR Gallup: BN = 0.613 vs ML = 0.850 — substantial gap,
#   partly attributable to evaluation asymmetry and the BN's
#   sparse structure (single parent for Life_Ladder).
# - Canadian Survey: BN = 0.076 vs ML = 0.423 — largest gap.
#   The restricted BN enforces a causally constrained structure
#   that excludes proxy predictors, accepting a large predictive
#   cost in exchange for interpretability.
#
# The BN vs ML gap is smallest where causal structure is cleanest
# (Student Depression, WHR Gallup unconstrained) and largest where
# expert constraints are strongest (Canadian Survey restricted BN).
# This directly illustrates the prediction-explanation trade-off:
# BNs optimized for causal plausibility sacrifice predictive
# accuracy, while ML models optimized for prediction may exploit
# causally ambiguous pathways.
#
# ### Key Cross-Dataset Finding
# Across all four datasets and all three model families, the
# primary driver of performance differences is data quality and
# signal concentration rather than model choice. XGBoost
# consistently leads or ties for best performance, but the
# differences between model families are small relative to
# the differences between datasets. This suggests that for
# wellbeing prediction, dataset selection and feature engineering
# matter more than algorithm choice — a practically important
# finding for applied researchers in this domain.
```
