```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch

fig, ax = plt.subplots(figsize=(12, 4))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis("off")

# Block positions and labels
blocks = [
    ("Data Sources", 0.05),
    ("Preprocessing\nLayer", 0.30),
    ("Modeling\nLayer", 0.55),
    ("Evaluation\nLayer", 0.80),
]

y = 0.4
w = 0.18
h = 0.3

# Draw boxes with better contrast
for label, x in blocks:
    box = FancyBboxPatch(
        (x, y),
        w,
        h,
        boxstyle="round,pad=0.05",
        linewidth=2,
        facecolor="#E6F0FF",   # light blue for visibility
        edgecolor="black"
    )
    ax.add_patch(box)

    ax.text(
        x + w / 2,
        y + h / 2,
        label,
        ha="center",
        va="center",
        fontsize=12,
        fontweight="bold",
        color="black"
    )

# Draw arrows
for i in range(len(blocks) - 1):
    x_start = blocks[i][1] + w
    x_end = blocks[i + 1][1]

    arrow = FancyArrowPatch(
        (x_start + 0.01, y + h / 2),
        (x_end - 0.01, y + h / 2),
        arrowstyle="->",
        mutation_scale=18,
        linewidth=2,
        color="black"
    )
    ax.add_patch(arrow)

plt.tight_layout()
plt.savefig("system_architecture_diagram.png", dpi=300, bbox_inches="tight")
plt.show()
```
